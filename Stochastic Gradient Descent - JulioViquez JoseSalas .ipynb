# Gradient Decent

Gradient Decent es un algoritmo de optimización que busca el minimizar una función. Para encontrar el mínimo local de una función, este algoritmo toma pasos siguiendo el lado mas negativo de la gradiente, en otras palabras, el algoritmo trata de descender siguiendo el camino más empinado de la función.
    
La siguiente ecuación describe el funcinamiento del algoritmo:
    
$$b = a - \gamma \nabla f(a)$$

$b$ representa la siguiente posición del algoritmo en la función de costo, $a$ representa la posición actual $\gamma$ es el factor de espera y $\nabla f(a)$ es la primer derivada de la función de costo e indica la dirección del descenso más empinado.

Ejemplo gráfico del funcionamiento del algoritmo con dos polinomios de diferente grado:

### Bibliotecas necesarias

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from sympy import diff, symbols, lambdify
```

### Definición de las funciones de costo

```python
#inicializacion de funciones de costo que se desean minimizar
x = symbols('x')

x2 = x**2             #Funcion cuadratica
x2_prime = diff(x2)   #Derivada de la funcion cuadratica

x5 = 1/20*(x+4)*(x+2)*(x+1)*(x-1)*(x-3)+2           #Funcion de un polinomio de grado 5
x5_prime = diff(x5)                                 #Derivada de la funcion de un polinomio de grado 5

x2 = lambdify(x, x2)               #Crea la funcion cuadratica
x2_prime = lambdify(x, x2_prime)   #Crea la funcion derivada de la funcion cuadratica

x5 = lambdify(x, x5)              #Crea la funcion de grado 5
x5_prime = lambdify(x, x5_prime)  #Crea la funcion derivada de la funcion de grado 5
```

### Algoritmo de Gradient Decent

```python
def gradient_descent(previous_x, learning_rate, epoch, cost_func, grad_function):
    x_gd = []
    y_gd = []
    
    x_gd.append(previous_x)
    y_gd.append(cost_func(previous_x))

    # Empieza el ciclo para actualizar x y y
    for i in range(epoch):
        current_x = previous_x - learning_rate * grad_function(previous_x)
        
        x_gd.append(current_x)
        y_gd.append(cost_func(current_x))

        # Actializa previous_x
        previous_x = current_x

    return x_gd, y_gd
```

### Ejemplo de Gradient Descent aplicado a una función cuadrática

```python
%matplotlib inline

#inicializacion de parametros para graficar la funcion cuadratica
x0 = -5             #punto de inicio
learning_rate = 0.15 #gamma
epoch = 10           #cantidad de iteraciones

x = np.arange(-5, 5, 0.01)  # genera los numeros entre -5 y 5

y2 = x2(x)
x2_gd, y2_gd = gradient_descent(x0, learning_rate, epoch, x2, x2_prime) #gradiente de la funcion cuadratica

y5 = x5(x)
x5_gd, y5_gd = gradient_descent(x0, learning_rate, epoch, x5, x5_prime) #gradiente de la funcion de grado 5

#grafica

plt.figure(1)
plt.scatter(x2_gd, y2_gd, c = 'b')
plt.plot(x, y2, c ='k')
for i in range(1, epoch+1):
    plt.annotate('', xy=(x2_gd[i], y2_gd[i]), xytext=(x2_gd[i-1], y2_gd[i-1]),
                   arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                   va='center', ha='center')

    plt.title('Min = ' + str(round(y2_gd[-1],7)))

plt.show()
```

### Ejemplo de Gradient Descent aplicado a una función de grado cinco

```python
#inicializacion de parametros para graficar la funcion del polinomio de grado 5
x0 = -3             #punto de inicio
learning_rate = 0.15 #gamma
epoch = 10           #cantidad de iteraciones

x = np.arange(-3, 3, 0.01)  # genera los numeros entre -5 y 5

y5 = x5(x)
x5_gd, y5_gd = gradient_descent(x0, learning_rate, epoch, x5, x5_prime) #gradiente de la funcion de grado 5

#grafica

plt.figure(1)
plt.scatter(x5_gd, y5_gd, c = 'b')
plt.plot(x, y5, c ='k')
for i in range(1, epoch+1):
    plt.annotate('', xy=(x5_gd[i], y5_gd[i]), xytext=(x5_gd[i-1], y5_gd[i-1]),
                   arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                   va='center', ha='center')
    
plt.title('Min = ' + str(round(y5_gd[-1],7)))

plt.show()
```

# Stochastic Gradient Descent



Stochastic Gradient Descent (SGD) consta de una variante del algoritmo de optimización Gradient Descent. Este algoritmo se encarga de encontrar el mínimo local de una función (por lo general la función de error) pero reduciendo el tamaño del lote por iteración significativamente. A diferencia del Gradient Descent, donde cada iteración se calculaba el error de todos los puntos del modelo, el Stochastic Gradient Descent calcula este error a partir de un subconjunto mínimo de puntos ahorrando de esta forma recursos computacionales. El SGD es utilizado principalmente cuando el dataset a procesar consta de millones de datos.

La siguiente ecuación describe el funcinamiento del algoritmo:
$$b = a - \gamma\nabla f( a; x^{(i)}; y^{(i)})$$

$b$ representa la siguiente posición del algoritmo en la función de costo, $a$ representa la posición actual $\gamma$ es el factor de espera y $f( a; x^{(i)}; y^{(i)})$ es la primer derivada de la función de costo e indica la dirección del descenso más empinado, en este caso $x^{(i)}; y^{(i)}$ representan la variable independiente y la variable dependiente respectivamente de un dataset.

La implementación de la primera derivada se realiza de la siguiente forma:
$$\frac{1}{2n} \sum_{i=0}^{n}[h{\Theta }(x^{i})-y^{i}]^{2}$$

Donde $h{\Theta }$ corresponde a la función de hipótesis y n es la cardinalidad del subconjunto de datos extraidos del dataset.


### Parámetros de la función

```python
m = 20 #cantidad de puntos de la función 
b = 8 #valor de b de la recta
theta1_true = -0.5 #pendiente de la recta
x = np.linspace(-1,1,m) #modelo en x
y = theta1_true * x + b #función objetivo
```

### Función de costo e hipótesis

```python
def funcion_costo(theta1):
    """
    Actualiza la función de costo con los nuevos valores de theta
    @param theta1: valor de la pendiente de la iteración actual, tipo lista
    @return Retorna el promedio de error con el theta actual
    """
    theta1 = np.atleast_2d(np.asarray(theta1))
    return np.average((y-hipotesis(x, theta1))**2, axis=1)/2

def hipotesis(x, theta1):
    """
    Crea una nueva función de hipótesis con el valor actual de tetha.
    @param x: tipo lista que contiene todos los puntos en x de la función objetivo
        tetha1:  valor de la pendiente de la iteración actual, tipo lista
    @return: retorna la función de hipótesis actualizada con la tetha del parámetro
    """
    return theta1*x + b
```

### Inicialización de funciones y parámetros

```python
theta1_grid = np.linspace(-1,1,m)
J_grid = funcion_costo(theta1_grid[:,np.newaxis])


N = 8 #Numero de iteraciones
alpha = 0.75 #learning rate
theta1 = [1.5]
J = [funcion_costo(theta1[0])[0]]
```

### Función de Stochastic Gradient Descent

```python
def stochastic_gradiant_descent():
    muestras = 5 #cardinalidad del subconjunto de datos a procesar
    for j in range(N-1):
        last_theta1 = theta1[-1]
        x1 = []
        y1 = []
        for t in range (0,muestras): #Se escoge aleatoriamente los datos del subconjunto
            random_index = random.randint(0,m-1)
            x1.append(x[random_index])
            y1.append(y[random_index])
        x1 = np.array(x1,np.float64)
        y1 = np.array(y1,np.float64)
        error_function = hipotesis(x1, last_theta1) - y1 #se calcula el error de esta iteración con el subconjunto anteriormente definido
        this_theta1 = last_theta1 - alpha / muestras * np.sum((error_function * x1)) #cálculo del siguiente punto en la función de costo
        theta1.append(this_theta1)
        J.append(funcion_costo(this_theta1))
```

### Representación Gráfica ejemplo 1:

```python
stochastic_gradiant_descent()

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6.15)) #Figura gráfica de la función de costo
ax[0].scatter(x, y, marker='*', s=30, color='k') #Marcas de la función objetivo
ax[1].plot(theta1_grid, J_grid, 'k')
#colores de las líneas y puntos por iteración
colors = ['b', 'g', 'm', 'c', 'orange','#2FE19F','#B7E131','#B70031','#4D0031','#3ECD00'] 
ax[0].plot(x, hipotesis(x, theta1[0]), color=colors[0], lw=2,label=r'$\theta_1 = {:.3f}$'.format(theta1[0]))

# Grafica de cada uno de las iteraciones
for j in range(1,N):
    ax[1].annotate('', xy=(theta1[j], J[j]), xytext=(theta1[j-1], J[j-1]),arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},va='center', ha='center')
    ax[0].plot(x, hipotesis(x, theta1[j]), color=colors[j], lw=2,label=r'$\theta_1 = {:.3f}$'.format(theta1[j]))

ax[1].scatter(theta1, J, c=colors, s=40, lw=0)
ax[1].set_xlim(-1,2)
ax[1].set_xlabel(r'$\theta_1$')
ax[1].set_ylabel(r'$J(\theta_1)$')
ax[1].set_title('Función de costo')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
ax[0].set_title('Modelo e hipótesis')
ax[0].legend(loc='upper left', fontsize='small')

plt.tight_layout()
plt.show()
```
